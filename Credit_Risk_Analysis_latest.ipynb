Now comes the exciting part‚Äîbuilding models to predict loan defaults. We‚Äôve chosen seven powerful algorithms, each bringing its own strengths to the table:

## üîç Investigating Outliers

Loop over columns to determine why an outlier exists:  

- **Mistake?** üìù (Data entry error, incorrect units, etc.)  
- **Genuine?** üìä (Extreme but valid value)  

‚û°Ô∏è **Is the outlier an error or a real anomaly?**  

```python
def detect_outliers(df, numerical_cols, lower_factor=1.5, upper_factor=1.5):
    outlier_data = pd.DataFrame()
    all_outliers_idx = set()  # Track indices of outliers

    for col in numerical_cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1

        lower_bound = Q1 - lower_factor * IQR
        upper_bound = Q3 + upper_factor * IQR

        if not df.empty:
            print(f"{col}: Min Value = {df[col].min()}, Max Value = {df[col].max()}")
            print(f"{col}: Lower Bound = {lower_bound}, Upper Bound = {upper_bound}\n")

        outliers = df[(df[col] > upper_bound) | (df[col] < lower_bound)]
        all_outliers_idx.update(outliers.index)  # Store outlier indices

    # Separate outliers and cleaned data
    outlier_data = df.loc[list(all_outliers_idx)]
    df_cleaned = df.drop(index=all_outliers_idx)

    print("Outliers detected and stored in 'outlier_data'. Remaining data cleaned.")

    return df_cleaned, outlier_data

```

```python
# Apply function
dfCleaned, outlierData = detect_outliers(df, numerical , 1.5 , 1.5)
```

```python
df['loan_status'] = pd.to_numeric(df['loan_status'], errors='coerce')

def bar_plot_custom(Attribute, hori=True, rot=False):
    """
    Plots two bar charts for a given categorical attribute from the credit risk dataset.

    Parameters:
    - Attribute (str): The column name in the dataframe to analyze.
    - hori (bool): If True, plots side-by-side horizontal subplots; if False, plots vertically.
    - rot (bool): If True, rotates the x-axis labels for readability.

    Assumes that the dataframe is named 'data' and the target column is 'loan_status'
    (1 indicates accepted loans and 0 indicates rejected loans).
    """

    # Calculate the normalized distribution of the attribute values
    variable = df[Attribute].value_counts(normalize=True)

    # Instead of aggregating the entire DataFrame, aggregate only the 'loan_status' column
    Variable_Data_plot = df.groupby(df[Attribute])['loan_status'].mean()

    # Plotting: side-by-side if hori=True, otherwise stacked vertically.
    if hori:
        plt.figure(figsize=[15,7])

        # Plot 1: Distribution of the categorical attribute
        plt.subplot(1,2,1)
        sns.barplot(x=variable.index, y=variable.values, palette='magma')
        plt.xlabel(Attribute)
        plt.ylabel("Percentage of Applicants")
        if rot:
            plt.xticks(rotation=90)

        # Plot 2: Average acceptance rate per category
        plt.subplot(1,2,2)
        sns.barplot(x=Variable_Data_plot.index, y=Variable_Data_plot.values, palette='magma')
        plt.xlabel(Attribute)
        plt.ylabel("Percentage of Accepted Loans")
        if rot:
            plt.xticks(rotation=90)

        plt.tight_layout()
        plt.show()
    else:
        plt.figure(figsize=[14,20])

        # Plot 1: Distribution of the categorical attribute (top)
        plt.subplot(2,1,1)
        sns.barplot(x=variable.index, y=variable.values, palette='magma')
        plt.xlabel(Attribute)
        plt.ylabel("Percentage of Applicants")
        if rot:
            plt.xticks(rotation=90)

        # Plot 2: Average acceptance rate per category (bottom)
        plt.subplot(2,1,2)
        sns.barplot(x=Variable_Data_plot.index, y=Variable_Data_plot.values, palette='magma')
        plt.xlabel(Attribute)
        plt.ylabel("Percentage of Accepted Loans")
        if rot:
            plt.xticks(rotation=90)

        plt.tight_layout()
        plt.show()
```

#### Outlier Treatment

```python
def remove_outliers_iqr(df, columns):
    """
    Removes rows containing outliers using the IQR method.
    """
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Filter out the outliers
        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

    return df

```

```python


# Define a function that computes all required metrics: accuracy, precision, recall, f1, and ROC AUC.
def get_all_metrics(y_test, y_pred, y_pred_prob):
    """
    Computes standard classification metrics along with the ROC AUC score.

    Parameters:
      y_test (array-like): True labels.
      y_pred (array-like): Predicted labels.
      y_pred_prob (array-like): Predicted probabilities (assumes second column corresponds to the positive class).

    Returns:
      A tuple containing:
         - Accuracy
         - Precision (weighted)
         - Recall (weighted)
         - F1 score (weighted)
         - ROC AUC score (for binary classification, using probability for the positive class)
    """
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')

    # For ROC AUC, we assume binary classification and that the probability of the positive class is in column 1.
    roc_auc = roc_auc_score(y_test, y_pred_prob[:, 1])
    return accuracy, precision, recall, f1, roc_auc

# Get predicted probabilities for Logistic Regression
y_pred_prob_log = logistic_model.predict_proba(X_test)

# Similar for other models (Decision Tree, Random Forest, etc.)
y_pred_prob_dt = dt_model.predict_proba(X_test)
y_pred_prob_rf = rf_model.predict_proba(X_test)
y_pred_prob_gb = gb_model.predict_proba(X_test)
y_pred_prob_knn = KNN_model.predict_proba(X_test)
y_pred_prob_svc = svc_model.predict_proba(X_test)  # Assuming SVC is configured for probability estimates
y_pred_prob_nb = nb_model.predict_proba(X_test)



log_metrics = get_all_metrics(y_test, y_pred_log, y_pred_prob_log)
dt_metrics = get_all_metrics(y_test, y_pred_dt, y_pred_prob_dt)
rf_metrics = get_all_metrics(y_test, y_pred_rf, y_pred_prob_rf)
gb_metrics = get_all_metrics(y_test, y_pred_gb, y_pred_prob_gb)
knn_metrics = get_all_metrics(y_test, y_pred_knn, y_pred_prob_knn)
svc_metrics = get_all_metrics(y_test, y_pred_svc, y_pred_prob_svc)
nb_metrics  = get_all_metrics(y_test, y_pred_nb, y_pred_prob_nb)

# Define a helper function to extract confusion matrix metrics (TP, FP, FN, TN) for binary classification.
def extract_confusion_metrics(conf_matrix):
    """
    Given a 2x2 confusion matrix, extracts:
      - True Positives (TP)
      - False Positives (FP)
      - False Negatives (FN)
      - True Negatives (TN)

    Note: This function only works for binary classification.
    """
    # For binary classification, the confusion matrix is 2x2.
    tn, fp, fn, tp = conf_matrix.ravel()
    return tp, fp, fn, tn

# Compute the confusion matrices for each model
conf_matrix_log = confusion_matrix(y_test, y_pred_log)
conf_matrix_dt  = confusion_matrix(y_test, y_pred_dt)
conf_matrix_rf  = confusion_matrix(y_test, y_pred_rf)
conf_matrix_gb  = confusion_matrix(y_test, y_pred_gb)
conf_matrix_knn = confusion_matrix(y_test, y_pred_knn)
conf_matrix_nb  = confusion_matrix(y_test, y_pred_nb)
conf_matrix_svc = confusion_matrix(y_test, y_pred_svc)

# Extract confusion metrics for each model
tp_log, fp_log, fn_log, tn_log = extract_confusion_metrics(conf_matrix_log)
tp_dt,  fp_dt,  fn_dt, tn_dt  = extract_confusion_metrics(conf_matrix_dt)
tp_rf,  fp_rf,  fn_rf, tn_rf  = extract_confusion_metrics(conf_matrix_rf)
tp_gb,  fp_gb,  fn_gb, tn_gb  = extract_confusion_metrics(conf_matrix_gb)
tp_knn, fp_knn, fn_knn, tn_knn = extract_confusion_metrics(conf_matrix_knn)
tp_nb,  fp_nb,  fn_nb, tn_nb  = extract_confusion_metrics(conf_matrix_nb)
tp_svc, fp_svc, fn_svc, tn_svc = extract_confusion_metrics(conf_matrix_svc)

# Now, merge all the metrics into one single DataFrame.
all_metrics_df = pd.DataFrame({
    'Model': [
        'Logistic Regression',
        'Decision Tree',
        'Random Forest',
        'Gradient Boosting',
        'K Nearest Neighbour',
        'Naive Bayes',
        'Support Vector Classifier'
    ],
    'Accuracy': [
        log_metrics[0], dt_metrics[0], rf_metrics[0], gb_metrics[0], knn_metrics[0], nb_metrics[0], svc_metrics[0]
    ],
    'Precision': [
        log_metrics[1], dt_metrics[1], rf_metrics[1], gb_metrics[1], knn_metrics[1], nb_metrics[1], svc_metrics[1]
    ],
    'Recall': [
        log_metrics[2], dt_metrics[2], rf_metrics[2], gb_metrics[2], knn_metrics[2], nb_metrics[2], svc_metrics[2]
    ],
    'F1 Score': [
        log_metrics[3], dt_metrics[3], rf_metrics[3], gb_metrics[3], knn_metrics[3], nb_metrics[3], svc_metrics[3]
    ],
    'ROC AUC': [
        log_metrics[4], dt_metrics[4], rf_metrics[4], gb_metrics[4], knn_metrics[4], nb_metrics[4], svc_metrics[4]
    ],
    'True Positives': [
        tp_log, tp_dt, tp_rf, tp_gb, tp_knn, tp_nb, tp_svc
    ],
    'False Positives': [
        fp_log, fp_dt, fp_rf, fp_gb, fp_knn, fp_nb, fp_svc
    ],
    'False Negatives': [
        fn_log, fn_dt, fn_rf, fn_gb, fn_knn, fn_nb, fn_svc
    ],
    'True Negatives': [
        tn_log, tn_dt, tn_rf, tn_gb, tn_knn, tn_nb, tn_svc
    ]
})

# Print the consolidated DataFrame containing all metrics
all_metrics_df
```

# Creating Pipeline

```python
import pandas as pd
import joblib

# Load the pre-trained pipeline
pipeline = joblib.load('pipeline.joblib')

def predict_applicant_risk(applicant_data):
    """
    Predicts the likelihood of loan approval and classifies the applicant risk.

    Parameters:
      applicant_data (pd.DataFrame): A single-row DataFrame containing applicant features.

    Returns:
      message (str): A message summarizing the applicant's risk profile and suggested action.
    """
    # Ensure the input is a DataFrame
    if not isinstance(applicant_data, pd.DataFrame):
        applicant_data = pd.DataFrame(applicant_data, index=[0])

    # Predict the approval probability.
    # Here, we assume that the positive class (e.g., approval) is at index 1.
    approval_probability = pipeline.predict_proba(applicant_data)[:, 1][0]

    # Define risk classification thresholds:
    if approval_probability >= 0.70:
        risk_level = "low risk"
        action = "proceed with the application."
    elif approval_probability >= 0.40:
        risk_level = "moderate risk"
        action = "further review is recommended before proceeding."
    else:
        risk_level = "high risk"
        action = "consider rejecting the application or requesting more information."

    # Format the message output
    message = (
        f"Applicant has a profile with {risk_level} "
        f"(predicted approval likelihood: {approval_probability*100:.1f}%). "
        f"Suggested action: {action}"
    )

    return message


new_applicant = {
    'person_age': 35,
    'person_income': 50000,
    'person_emp_length': 10,
    'loan_amnt': 15000,
    'loan_int_rate': 12.5,
    'loan_percent_income': 30,
    'cb_person_cred_hist_length': 5,
    'person_home_ownership': 'OWN',
    'loan_intent': 'credit_card',
    'loan_grade': 'B',
    'cb_person_default_on_file': 'N'
}


applicant_df = pd.DataFrame([new_applicant])

# Get the prediction message.
result_message = predict_applicant_risk(applicant_df)
print(result_message)
```
